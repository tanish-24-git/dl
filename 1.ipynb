```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\pra\Datasets\1_iris.csv")
```

```python
# 3. Features & Labels
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

```

```python

# 4. Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# 5. Model (KNN)
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

```python
# 6. Prediction
pred = model.predict(X_test)

# 7. Accuracy
print("Accuracy:", accuracy_score(y_test, pred))
```

```python
# 8. Visualize patterns (Sepal Length vs Sepal Width)
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=pd.factorize(y)[0])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Pattern Recognition")
plt.show()
```
