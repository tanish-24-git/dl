```python
# 1. Import
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score

# 2. Load
train = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\pra\Datasets\4_mnist_train.csv")
test = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\pra\Datasets\4_mnist_test.csv")

X_train = train.iloc[:,1:]
y_train = train.iloc[:,0]

X_test = test.iloc[:,1:]
y_test = test.iloc[:,0]

# 3. Scale (IMPORTANT)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 4. Model
model = MultinomialNB()
model.fit(X_train, y_train)

# 5. Predict
pred = model.predict(X_test)

# 6. Accuracy
print("Accuracy:", accuracy_score(y_test, pred))
```

```python
import matplotlib.pyplot as plt

# Show one image
img = X_test[0].reshape(28,28)

plt.imshow(img, cmap='gray')
plt.title("Actual: " + str(y_test.iloc[0]) + 
          " | Pred: " + str(pred[0]))
plt.axis('off')
plt.show()
```
