```python
import numpy as np

# Inputs
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Weights (x1 positive, x2 inhibitory)
w1 = 1
w2 = -1
threshold = 1

print("x1 x2 -> Output")

for x in X:
    net = x[0]*w1 + x[1]*w2
    
    # Step activation
    if net >= threshold:
        y = 1
    else:
        y = 0
        
    print(x[0], x[1], "->", y)
```
