```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam

# Load dataset
data = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\aissms-clg-works\DL Dataset\DL Dataset\Pr-10\mnist_train.csv")

# Normalize images
X = data.iloc[:, 1:].values / 255.0

# Add noise
noise = 0.5 * np.random.normal(size=X.shape)
X_noisy = X + noise
X_noisy = np.clip(X_noisy, 0., 1.)

# Autoencoder
input_img = Input(shape=(784,))

# Encoder (compression)
encoded = Dense(128, activation='relu')(input_img)
encoded = Dense(64, activation='relu')(encoded)

# Decoder (reconstruction)
decoded = Dense(128, activation='relu')(encoded)
decoded = Dense(784, activation='sigmoid')(decoded)

# Model
autoencoder = Model(input_img, decoded)

# Compile
autoencoder.compile(optimizer=Adam(), loss='mse')

# Train
autoencoder.fit(X_noisy, X, epochs=5, batch_size=256)

# Predict (denoise)
decoded_imgs = autoencoder.predict(X_noisy[:5])

print("Denoised output generated")
```
