```python
import numpy as np
import pandas as pd

# Load dataset (make sure file is in same folder)
data = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\aissms-clg-works\DL Dataset\DL Dataset\Pr-04\xor_data.csv")

# Assuming columns: x1, x2, y
X = data.iloc[:, 0:2].values
y = data.iloc[:, 2].values.reshape(-1,1)

# Activation
def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_deriv(x):
    return x*(1-x)

# Initialize weights
np.random.seed(1)
w1 = np.random.rand(2,2)
b1 = np.zeros((1,2))

w2 = np.random.rand(2,1)
b2 = np.zeros((1,1))

lr = 0.5

# Training
for epoch in range(10000):

    # Forward
    h = sigmoid(np.dot(X, w1) + b1)
    output = sigmoid(np.dot(h, w2) + b2)

    # Error
    error = y - output

    # Backprop
    d_output = error * sigmoid_deriv(output)
    d_hidden = d_output.dot(w2.T) * sigmoid_deriv(h)

    # Update
    w2 += h.T.dot(d_output) * lr
    b2 += np.sum(d_output, axis=0, keepdims=True) * lr

    w1 += X.T.dot(d_hidden) * lr
    b1 += np.sum(d_hidden, axis=0, keepdims=True) * lr

# Final Output
print("Predicted Output:")
print(np.round(output, 2))
```
