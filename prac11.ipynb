```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, LeakyReLU, Input
from tensorflow.keras.optimizers import Adam

data = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\aissms-clg-works\DL Dataset\DL Dataset\Pr-11\mnist_train.csv")
X = data.iloc[:, 1:].values / 255.0

latent_dim = 100

generator = Sequential([Dense(128, input_dim=latent_dim), LeakyReLU(0.2), Dense(784, activation='sigmoid')])
discriminator = Sequential([Dense(128, input_dim=784), LeakyReLU(0.2), Dense(1, activation='sigmoid')])
discriminator.compile(loss='binary_crossentropy', optimizer=Adam(0.0002, 0.5), metrics=['accuracy'])

z_input = Input(shape=(latent_dim,))
fake_img = generator(z_input)
discriminator.trainable = False
gan = Model(z_input, discriminator(fake_img))
gan.compile(loss='binary_crossentropy', optimizer=Adam(0.0002, 0.5))

for epoch in range(10):
    idx = np.random.randint(0, X.shape[0], 32)
    real = X[idx]
    discriminator.train_on_batch(real, np.ones((32, 1)))
    
    noise = np.random.normal(0, 1, (32, latent_dim))
    fake = generator.predict(noise, verbose=0)
    discriminator.train_on_batch(fake, np.zeros((32, 1)))
    
    gan.train_on_batch(np.random.normal(0, 1, (32, latent_dim)), np.ones((32, 1)))
    
    if (epoch + 1) % 5 == 0:
        print(f"Epoch {epoch+1}/10")

noise = np.random.normal(0, 1, (1, latent_dim))
gen_img = generator.predict(noise).reshape(28, 28)
print("Generated image shape:", gen_img.shape)
```
