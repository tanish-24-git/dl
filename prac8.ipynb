```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load dataset
data = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\aissms-clg-works\DL Dataset\DL Dataset\Pr-08\IMDB Dataset.csv")

# Text & labels
X = data['review']
y = data['sentiment'].map({'positive':1, 'negative':0})

# Tokenization
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(X)
X_seq = tokenizer.texts_to_sequences(X)

# Padding
X_pad = pad_sequences(X_seq, maxlen=200)

# Split
X_train, X_test, y_train, y_test = train_test_split(X_pad, y, test_size=0.2)

# Model (RNN)
model = Sequential()
model.add(Embedding(5000, 32, input_length=200))
model.add(SimpleRNN(32))
model.add(Dense(1, activation='sigmoid'))

# Compile
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# Train
model.fit(X_train, y_train, epochs=3, verbose=1)

# Evaluate
loss, acc = model.evaluate(X_test, y_test)
print("Accuracy:", acc)
```
