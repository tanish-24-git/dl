```python
# 1. Import
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# 2. Load Dataset
data = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\pra\Datasets\3_winequality-red.csv")

# 3. Features & Label
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

# 4. Convert to Classification (optional: quality high/low)
y = (y >= 7).astype(int)

# 5. Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 6. Model (Gaussian NB)
model = GaussianNB()
model.fit(X_train, y_train)

# 7. Prediction
pred = model.predict(X_test)

# 8. Accuracy
print("Accuracy:", accuracy_score(y_test, pred))
```
