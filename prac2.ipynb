```python
import numpy as np
import matplotlib.pyplot as plt

# Data
X_raw = np.arange(0, 95)

# Features: number + modulo feature
X = np.column_stack((X_raw/100.0, X_raw % 2))

# Labels: Even = -1, Odd = +1
y = np.array([1 if i % 2 else -1 for i in X_raw])
```

```python
w = np.zeros(2)
b = 0
lr = 0.1

for epoch in range(20):
    for i in range(len(X)):
        net = np.dot(w, X[i]) + b
        y_pred = 1 if net >= 0 else -1

        if y[i] * y_pred <= 0:
            w = w + lr * y[i] * X[i]
            b = b + lr * y[i]

print("w:", w)
print("b:", b)
```

```python
test = np.array([10, 15, 22, 33])
test_X = np.column_stack((test/100.0, test % 2))

for val, x in zip(test, test_X):
    net = np.dot(w, x) + b
    pred = 1 if net >= 0 else -1
    print(val, "->", "Odd" if pred==1 else "Even")
```

```python
plt.scatter(X[:,0], y, c=y)
plt.title("Even (-1) vs Odd (+1)")
plt.xlabel("Number")
plt.ylabel("Class")
plt.grid()
plt.show()
```
